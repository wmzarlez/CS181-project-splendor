```python
import numpy as np
import math
R = 0.5
```

```python
def print_element(pos, n, texCord):
    print("Element{.pos = {%f, %f, %f}, .n = {%f, %f, %f}, .text_cord = {%f, %f}}, "%(pos[0], pos[1], pos[2], n[0], n[1], n[2], texCord[0], texCord[1]))
```

```python
# generate the front surface elements of the coin
for i in range(0, 360, 10):
    theta =  math.radians(i)
    print_element((math.cos(theta), math.sin(theta), 0.1), (0.0, 0.0, 1.0), (0.5 + 0.5 * math.cos(theta), 0.5 + 0.5 * math.sin(theta)))
```

```python
#generate the indices of the front surface
for i in range(1, 37, 1):
    print("%u, %u, %u, "%(0, i, (i) % 36 + 1))
```

```python
# generate the front surface elements of the coin
for i in range(0, 360, 10):
    theta =  math.radians(i)
    print_element((math.cos(theta), math.sin(theta), -0.1), (0.0, 0.0, -1.0), (0.5 - 0.5 * math.cos(theta), 0.5 + 0.5 * math.sin(theta)))
```

```python
#generate side surface
for i in range(0, 360, 10):
    theta = math.radians(i)
    print_element((math.cos(theta), math.sin(theta), 0.1), (math.cos(theta), math.sin(theta), 0), (-1, -1))
    print_element((math.cos(theta), math.sin(theta), -0.1), (math.cos(theta), math.sin(theta), 0), (-1, -1))
```
